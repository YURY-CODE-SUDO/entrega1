##01 Visualizaciones univariadas








Estadística y calculo del género que en ventas globales más vendio y el que menos vendio

```python
# Agrupar por género y sumar las ventas globales
genre_sales = video_games_df.groupby('Genre')['Global_Sales'].sum().sort_values(ascending=False)

# Género más vendido
most_sold_genre = genre_sales.idxmax()
most_sold_value = genre_sales.max()

# Género menos vendido
least_sold_genre = genre_sales.idxmin()
least_sold_value = genre_sales.min()

# Resultados
print(f"Género más vendido: {most_sold_genre} con {most_sold_value:.2f} millones de ventas.")
print(f"Género menos vendido: {least_sold_genre} con {least_sold_value:.2f} millones de ventas.")
```

Ventas globales por género

```python
# Gráfico: Ventas globales por género
plt.figure(figsize=(12, 6))
sns.barplot(x='Global_Sales', y='Genre', data=video_games_df, ci=None, palette="viridis", estimator=sum)
plt.title('Ventas Globales por Género', fontsize=16)
plt.xlabel('Ventas Globales (en millones)', fontsize=12)
plt.ylabel('Género', fontsize=12)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.countplot(x='Genre', data=video_games_df, order=video_games_df['Genre'].value_counts().index, palette="viridis")
plt.title('Distribución de Videojuegos por Género', fontsize=16)
plt.xlabel('Género', fontsize=12)
plt.ylabel('Cantidad de Videojuegos', fontsize=12)
plt.xticks(rotation=45)  # Rotar etiquetas si es necesario
plt.show()
```

```python
# Agrupar por género y calcular estadísticas
critic_score_by_genre = video_games_cleaned.groupby('Genre')['Critic_Score'].agg(['mean', 'median', 'count']).reset_index()

# Ordenar por puntaje promedio
critic_score_by_genre_sorted = critic_score_by_genre.sort_values(by='mean', ascending=False)
print(critic_score_by_genre_sorted)
```

##Ventas Globales por Platafroma

Calculo de la plataforma que más vendió y la que menos vendió

```python
import pandas as pd

# Agrupar por plataforma y calcular la suma de las ventas globales
ventas_plataforma = video_games_df.groupby("Platform")["Global_Sales"].sum().reset_index()

# Ordenar de mayor a menor
ventas_plataforma = ventas_plataforma.sort_values("Global_Sales", ascending=False)

# Plataforma con más ventas
plataforma_mas_ventas = ventas_plataforma.iloc[0]

# Plataforma con menos ventas
plataforma_menos_ventas = ventas_plataforma.iloc[-1]

# Mostrar resultados
print(f"La plataforma que más vendió videojuegos es: {plataforma_mas_ventas['Platform']} con {plataforma_mas_ventas['Global_Sales']} millones de unidades.")
print(f"La plataforma que menos vendió videojuegos es: {plataforma_menos_ventas['Platform']} con {plataforma_menos_ventas['Global_Sales']} millones de unidades.")
```
